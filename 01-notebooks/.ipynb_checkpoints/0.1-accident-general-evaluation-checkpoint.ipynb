## 3. Evaluation of the influence of diferent features causing accidents

In this evaluation we focus on the data that doesn't have geolocation and try to explore the impact that different features have on the rate of accidents. For now we focus on the hour, meteorological conditions and other factors.

### 3.1. Evaluate the hour of the day

Here we try to evaluate the variance of accident rates over the hour of the day. Do accidents occur at specific times of the day?

```python
hour_counting = (
    data_wo_geolocation
    .groupby(['hour', 'accident_type'])
    .agg(percentage = ('date', np.size))
)
plt.figure(figsize = (25, 6))
sns.lineplot(data = hour_counting, x='hour', y='percentage', hue='accident_type', marker='o')
```

Main points:
1. Has we have seen before most of the accidents are colisions;
2. There seems to be two main peaks, one arroudn 9 o'clock and another arround 18 o'clock;
3. These times are the rush hours and so we normally have more cars moving and it makes sense that there are more accidents;
4. To have a better look of both type of accidents lets normalize the data.

```python
hour_counting = hour_counting.groupby('accident_type').apply(lambda x: 100 * x / float(x.sum())).reset_index()
plt.figure(figsize = (25, 6))
sns.lineplot(data = hour_counting, x='hour', y='percentage', hue='accident_type', marker='o')
```

Main points:
1. After the normalization we can see that both types of accidents have the same trends;
2. However, we also know that the hour of the day is probably a confounding variable and we will need to evaluate traffic levels in the city to understantd the impact it has n the accidents.

### 3.2. Evaluate the day of the week

Ok now that we evaluated how the number of accidents varies over time, we ave some ideias when more accidents occur. However, does this rate occurs every day of the week? Or are there some variations over the days of the week?

```python
weekday_counting = (
    data_wo_geolocation
    .groupby(['weekday', 'accident_type'])
    .agg(percentage = ('date', np.size))
)
plt.figure(figsize = (25, 6))
sns.lineplot(data = weekday_counting, x='weekday', y='percentage', hue='accident_type', marker='o')
```

Main points:
1. Has expected the day with less accidents is Saturday and Sunday (5 and 6), however only slightly;
2. Aparentyly monday and tuesday also have lower number of accidents, what could explain this?
3. Once again lets normalize the data and reevaluate the graph;

```python
weekday_counting = weekday_counting.groupby('accident_type').apply(lambda x: 100 * x / float(x.sum())).reset_index()
plt.figure(figsize = (25, 6))
sns.lineplot(data = weekday_counting, x='weekday', y='percentage', hue='accident_type', marker='o')
```

Main points:
1. It seems that the run overs have a more stable number of occurences over the week and a big decline in the weekend. 
2. Again, is this due to the smaller number of cars that are in the city?
3. The colisions continue to have the same behaviour with the lower number in mondays and tuesdays (0 and 1) and Sundays and Saturdays (5 and 6);

### 3.3. Evaluate the month
Another time feature we can evaluate is the number of accidents per months. Using the previous information we would expect to see a smaller number of accidnets during the summer months when most of the people take vacations. at the same time these are probably the times when most of the tourists come to visit Lisbon... Lets explore the data.

```python
month_counting = data_wo_geolocation.groupby(['month', 'accident_type']).agg(percentage = ('date', np.size))
plt.figure(figsize = (25, 6))
sns.lineplot(data = month_counting, x='month', y='percentage', hue='accident_type', marker='o')
```

Main points:
1. The months with less number of accidents seem to be February and June. Could this be an influence of the number of days per month?
2. The month with a bigger number of accidents is November.
3. Once again we have to normalize the data to better compare the two  types of accidents. However, we have to normalize per number of days and then convert to percentage;

```python
month_counting = (
    month_counting
    .reset_index()
    .assign(mes = lambda x: pd.to_datetime(x['month'].astype(str) + '-01-2019'))
    .assign(n_dias = lambda x: x['mes'].dt.daysinmonth)
    .assign(percentage = lambda x: x['percentage'] / x['n_dias'])
    .drop(columns = ['n_dias', 'mes'])
    .set_index(['month', 'accident_type'])
    .groupby('accident_type')
    .apply(lambda x: 100 * x / float(x.sum()))
    .reset_index()
)
plt.figure(figsize = (25, 6))
sns.lineplot(data = month_counting, x='month', y='percentage', hue='accident_type', marker='o')
```

Main points:
1. The month with less colisions is June and the one with less run overs is August;
2. The month with more accidents was november, however we can see a great increment of accidents in the end of the year;

### 3.6. Evaluate the influence of rain in the collisions

**`Get and clean rain data`**

```python
rain = pd.read_csv('../00-data/external/POWER_Regional_Daily_20190101_20191231_9c24c3cf.csv', sep=',', skiprows=19)
```

```python
clean_rain = (
    rain
    .assign(date = lambda x: pd.to_datetime(x['YEAR'].astype(str) + '-' + x['MO'].astype(str) + '-' + x['DY'].astype(str)))
    .assign(date_str = lambda x: x['date'].dt.strftime('%Y%m%d'))
    [['PRECTOT', 'date_str']]
    .rename(columns = {'PRECTOT': 'precipitation', 
                       'date_str': 'date'})
)
```

**`Add the data to the accidents data`**

```python
data_wo_geolocation_w_rain = (
    data_wo_geolocation
    .assign(date = lambda x: x.date.dt.strftime('%Y%m%d'))
    .groupby(['date', 'accident_type']).agg(n_accidents = ('accident_type', np.size))
    .reset_index()
    .merge(clean_rain, on = 'date')
    .pivot(index = 'date', columns = 'accident_type', values = ['n_accidents', 'precipitation'])
)
```

```python
data_wo_geolocation_w_rain.corr(method='spearman')
```

Main points:
1. The correlation of precipitation and number of colisions is small;
2. The correlation of the precipitation and the run overs is pratically null;
3. However, the correlation for colision sis still positive and could be a factor to be considered.

### 3.7. Evaluate the influence of night and day

**`Get and clean night day data`**

```python
day_night = pd.read_excel('../00-data//external/Por do sol.xlsx')
```

```python
clean_day_night = (
    day_night
    .assign(sunrise_time = lambda x: '2019-' + x['Month'].astype(str) + '-' + x['Day'].astype(str) + ' ' + x['Sunrise'].str.split().str[0])
    .assign(sunrise_time = lambda x: pd.to_datetime(x['sunrise_time'], format='%Y-%m-%d %H:%M'))
    .drop(columns = 'Sunrise')
    .assign(sunset_time = lambda x: '2019-' + x['Month'].astype(str) + '-' + x['Day'].astype(str) + ' ' + x['Sunset'].str.split().str[0])
    .assign(sunset_time = lambda x: pd.to_datetime(x['sunset_time'], format='%Y-%m-%d %H:%M'))
    .drop(columns = 'Sunset')
    .assign(date = lambda x: pd.to_datetime('2019-' + x['Month'].astype(str) + '-' + x['Day'].astype(str)).dt.strftime('%Y%m%d'))
    .drop(columns = ['Month', 'Day'])
)
```

**`Add the data to the accidents data`**

```python
def night_day(df):
    df['day_night'] = 'day'
    df.loc[(df['complete_date'] < df['sunrise_time']) |
           (df['complete_date'] > df['sunset_time']), 'day_night'] = 'night'
    return df
```

```python
data_wo_geolocation_w_day_night = (
    data_wo_geolocation
    .assign(date_str = lambda x: x['date'].dt.strftime('%Y%m%d'))
    .rename(columns = {'date': 'complete_date',
                       'date_str': 'date'})
    .merge(clean_day_night, on = 'date')
    .pipe(night_day)
)
```

```python
data_wo_geolocation_w_day_night.day_night.value_counts()
```

Main ponints:
1. Most of the accidents happend during the day, but it is also during that time that there is more movement;
2. Probably there is a great deal of confounding variables so we need a more robust evaluation.
